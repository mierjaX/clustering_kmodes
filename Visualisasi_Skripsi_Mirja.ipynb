```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Load the Excel file
file_path = 'cluster 1.xlsx'  # Path to the uploaded file
df = pd.read_excel(file_path, engine='openpyxl')

# Extract the relevant data
categories = df.columns[1:]  # Skip the first column which is 'Rating'
ratings = df[df.columns[0]]  # First column is assumed to be the 'Rating'

# Reorder the data based on the rating categories from "Sangat Buruk" to "Sangat Baik"
rating_order = ["Sangat Buruk", "Buruk", "Cukup", "Baik", "Sangat Baik"]
df['Unnamed: 0'] = pd.Categorical(df['Unnamed: 0'], categories=rating_order, ordered=True)
df = df.sort_values('Unnamed: 0')  # Sort by the new order

# Reverse the order of categories (so 'rating_overall' is at the top and 'harga' at the bottom)
categories = categories[::-1]

# Define the colors in the correct order (from "Sangat Buruk" to "Sangat Baik")
colors = ['#D32F2F', '#F57C00', '#FBC02D', '#388E3C', "#00796B" ]
rating_labels = ["Sangat Buruk", "Buruk", "Cukup", "Baik", "Sangat Baik"]

# Initialize the figure
fig, ax = plt.subplots(figsize=(14, 5))

# Loop through the ratings and plot the stack for each category
cumulative_values = [0] * len(categories)  # To track cumulative values for stacking

# Plot the bars for each rating category in the correct order
for idx, rating in enumerate(rating_order):
    values = df[df['Unnamed: 0'] == rating][categories].values.flatten()
    bars = ax.barh(categories, values, left=cumulative_values, color=colors[idx], label=rating, height=0.6,)

    # Update the cumulative values for stacking
    cumulative_values = [i + j for i, j in zip(cumulative_values, values)]

# Create the legend
legend_patches = [mpatches.Patch(color=colors[i], label=rating_labels[i]) for i in range(len(rating_labels))]
plt.legend(handles=legend_patches, bbox_to_anchor=(0.5, 1.06), loc='lower center', ncol=len(rating_labels), fontsize=12)

# Customize the chart
ax.set_xlabel('Jumlah Anggota', fontsize=12)
ax.set_ylabel('Atribut Kepuasan Pelanggan', fontsize=12)
ax.set_title('Cluster 1', fontsize=16)

# Adjust the layout to fit everything
plt.xticks(fontsize=12)  # Adjust x-axis font size
plt.yticks(fontsize=12)  # Adjust y-axis font size

# Adjust the layout to fit everything
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplots_adjust(top=1.2)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

# Load the Excel file
file_path = 'Data Survey Setelah Preprocessing Visualisasi.xlsx'  # Replace this with your actual file path
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Define the category order on the x-axis
category_order = ["15 - 25 Tahun", "26 - 35 Tahun", "36 - 45 Tahun", "46 - 56 Tahun", "Diatas 57 Tahun"]

# Assign custom colors for each category
colors = {
    "15 - 25 Tahun": "#D32F2F",
    "26 - 35 Tahun": "#F57C00",
    "36 - 45 Tahun": "#FBC02D",
    "46 - 56 Tahun": "#388E3C",
    "Diatas 57 Tahun": "#00796B"
}

# Set up the figure for subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 12), sharey=True)

# Loop through each cluster (1, 2, and 3) and create a subplot for each
for cluster, ax in zip([1, 2, 3], axes):
    # Filter data for the specific cluster
    cluster_data = data[data['Cluster'] == cluster]

    # Create a countplot for 'usia' for each cluster
    plot = sns.countplot(
        data=cluster_data,
        x='usia',
        order=category_order,  # ensure the same order for all subplots
        ax=ax,
        palette=colors  # apply custom colors for each category
    )

    # Add the exact count labels on top of each bar
    for p in plot.patches:
        plot.annotate(format(p.get_height(), '.0f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center',
                      xytext=(0, 9),  # Adjust label position
                      fontsize=16,  # Adjust label font size
                      textcoords='offset points')

    # Set titles and labels
    ax.set_title(f'Cluster {cluster}', fontsize=18)
    ax.set_xlabel('')

    # Rotate the x-axis labels to be vertical
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=16)  # Rotate x-axis labels

    ax.set_ylabel('Jumlah Anggota' if cluster == 1 else '', fontsize=16)  # Increase font size for y-axis label
    ax.tick_params(axis='y', labelsize=16)

# Create custom legend patches for categories based on colors
legend_patches = [mpatches.Patch(color=colors[cat], label=cat) for cat in category_order]

# Adjust the layout and add a common legend at the top
fig.legend(handles=legend_patches, loc='upper center', fontsize=16, title_fontsize=18, ncol=5, bbox_to_anchor=(0.5, 0.94))  # Adjust bbox_to_anchor to move the legend

# # Add overall title with adjusted position to avoid overlapping with the legend
# plt.suptitle('Usia', fontsize=20, y=0.94)  # Lower the title to avoid overlapping

# Adjust layout to ensure proper spacing for the legend on top
plt.tight_layout(rect=[0, 0, 1, 0.90])  # Adjust the 'rect' to create space between the legend and title
plt.show()
```
